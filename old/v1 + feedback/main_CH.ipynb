```python
# import packages
import pandas as pd
import requests
from pprint import pprint
```

```python
# get response from api request
response = requests.get(api)
display(response)
print(response.url)
```

```python
player_dict = response.json()
pprint(player_dict['1264'])

with open("players.json", "w+") as fo:
    print(response.text, file=fo)
```

```python
# CH
def draft(player_name, player_dict, user_draft):
    '''finds player and add to list of user's players
    params: player_name (string)
            player_dict (dict)
            user_draft (list)
    returns: user_draft
    '''
    found_player = False
    for key in player_dict:
        player = player_dict[key]
        full_name = player['first_name'] + " " + player['last_name']
        if full_name == player_name:
            user_draft.append(player)
            print(f"you have drafted {player_name}")
            found_player = True

    if found_player == False:
        print(f"player {player_name} not found")

    return user_draft
```

```python
'''
def calculate_grade calculates draft grade by predicted points

params: list user_draft of players
returns: grade
'''
def calculate_grade(user_draft):
    score = 0
    for key in user_draft:
        player = user_draft[key]
        score += player['search_rank']

    if score <= 300:
        return "S"
    if score <= 325:
        return "A"
    if score <= 350:
        return "B"
    if score <= 375:
        return "C"
    if score <= 400:
        return "D"
    if score <= 425:
        return "F"
```

```python
# function to read in players from json file
import json
def read_players(filename):
    with open(filename , "r") as fo:
        player_dict = json.load(fo) 
    return player_dict
```

https://docs.sleeper.com/#players

```python

user_draft = []
player_dict = read_players("players.json")
while True:
    player_name = input("max player size is 5. enter player's full name here: ")

    user_draft = draft(player_name, player_dict, user_draft)

    if len(user_draft) > 5:
        break

print("your draft rating is a: " + calculate_grade(user_draft))
```
